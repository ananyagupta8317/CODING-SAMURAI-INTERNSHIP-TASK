```python
import pandas as pd

data = pd.read_csv('IBM Dataset.csv')
print(data)
```

```python
# Calculate the attrition rate

attrition_rate = (data['Attrition'] == 'Yes').mean()*100
print("Attrition Rate:", attrition_rate)
```

```python
# Visualize attrition by department using a bar chart

attrition_department = pd.crosstab(data['Department'],data['Attrition'])
print(attrition_department)

import matplotlib.pyplot as plt
attrition_department.plot(kind ='bar')
plt.title("Attrition by Department")
plt.xlabel("Department")
plt.ylabel("Employee Count")
plt.show()
```

```python
# Identify the average age of employees leaving the company

Avg_age = data[data["Attrition"] == "Yes"]['Age'].mean()
print("Average age of employee who left the company:",Avg_age)
```
